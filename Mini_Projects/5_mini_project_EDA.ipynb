ðŸ”¹ 1. Load the Dataset

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv')
df.head()
```

ðŸ”¹ 2. Q: Whatâ€™s the average tip by gender?

```python
avg_tip_gender = df.groupby("sex")["tip"].mean()
print(avg_tip_gender)

# Bar chart
avg_tip_gender.plot(kind="bar", color=["skyblue", "pink"])
plt.title("Average Tip by Gender")
plt.ylabel("Tip ($)")
plt.show()
```

ðŸ”¹ 3. Q: Do smokers tip more or less than non-smokers?

```python
avg_tip_smoker = df.groupby("smoker")["tip"].mean()
print(avg_tip_smoker)

# Bar chart
avg_tip_smoker.plot(kind="bar", color=["gray", "green"])
plt.title("Average Tip: Smokers vs Non-Smokers")
plt.ylabel("Tip ($)")
plt.show()
```

ðŸ”¹ 4. Q: What day has the highest average total bill?


```python
avg_total_per_day = df.groupby("day")["total_bill"].mean()
print(avg_total_per_day)

# Bar chart
avg_total_per_day.plot(kind="bar", color="orange")
plt.title("Average Total Bill by Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

ðŸ”¹ 5. Create a Histogram of Tips

```python
plt.hist(df["tip"], bins=20, color="purple", edgecolor="black")
plt.title("Histogram of Tip Amounts")
plt.xlabel("Tip ($)")
plt.ylabel("Frequency")
plt.show()
```

ðŸ”¹ 6. Create a Line Plot of Total Bill Over Index


```python
plt.plot(df["total_bill"], color="teal")
plt.title("Total Bill Over Time (Index)")
plt.xlabel("Index")
plt.ylabel("Total Bill ($)")
plt.show()
```

ðŸ“ Summary
Males tipped more on average: Yes, males tipped slightly more than females.
Smokers tipped less than non-smokers.
Saturday had the highest average total bill.
Most tips were between $2 and $4 based on the histogram.
The line plot shows that total bills vary but generally range from $10 to $50.
